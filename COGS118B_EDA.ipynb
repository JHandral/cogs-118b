```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
data = pd.read_csv('features_30_sec.csv')
```

```python
data.head()
```

```python
data.shape
```

```python
plt.figure(figsize=(12, 8), dpi=80)
sns.scatterplot(data=data,x='spectral_centroid_mean',y='spectral_bandwidth_mean',hue='label',legend=True,
               alpha=0.7)
plt.legend(loc='lower right')
plt.title('Spectral Centroid Mean vs Spectral Bandwidth Mean by Genre')
plt.show()
```

```python
plt.figure(figsize=(12, 8), dpi=80)
sns.scatterplot(data=data[data['label'].isin(['classical','hiphop'])],
                x='spectral_centroid_mean',y='spectral_bandwidth_mean',hue='label',legend=True,
               alpha=0.8)
plt.legend(loc='lower right')
plt.title('Spectral Centroid Mean vs Spectral Bandwidth Mean - Classical vs Hiphop')
plt.show()a
```

```python
plt.figure(figsize=(12, 8), dpi=80)
sns.boxplot(data=data,x='label',y='tempo',
           order=['reggae','classical','blues','metal','disco','rock','country','jazz','pop','hiphop'])
plt.title('Tempo Distribution by Genre')
plt.show()
```

```python
plt.figure(figsize=(12, 8), dpi=80)
sns.boxplot(data=data,x='label',y='harmony_mean')
plt.title('Harmony Mean Distribution by Genre')
plt.show()
```

```python
plt.figure(figsize=(12, 8), dpi=80)
sns.boxplot(data=data,x='label',y='zero_crossing_rate_mean',
           order=['metal','pop','disco','rock','hiphop','reggae','country','classical','blues','jazz'])
plt.title('Zero Crossing Rate Mean Distribution by Genre')
plt.show()
```

```python
plt.figure(figsize=(12, 8), dpi=80)
sns.scatterplot(data=data,x='perceptr_var',y='perceptr_mean',hue='label',legend=True,
               alpha=0.7)
plt.legend(loc='lower right')
plt.title('Perceptrual Variance vs Perceptrual Mean by Genre')
plt.show()
```

```python
plt.figure(figsize=(12, 8), dpi=80)
sns.boxplot(data=data,x='label',y='spectral_centroid_mean',
           order=['pop','metal','disco','hiphop','rock','reggae','country','jazz','blues','classical'])
plt.title('Spectral Centroid Mean by Genre')
plt.show()
```

```python
plt.figure(figsize=(12, 8), dpi=80)
sns.boxplot(data=data,x='label',y='rolloff_mean',
           order=['pop','disco','hiphop','metal','rock','reggae','country','blues','jazz','classical'])
plt.title('Spectral Rolloff Mean by Genre')
plt.show()
```

```python
plt.figure(figsize=(12, 8), dpi=80)
sns.scatterplot(data=data[data['label'].isin(['classical','pop'])],
                x='spectral_centroid_mean',y='rolloff_mean',hue='label',legend=True,
               alpha=0.8)
plt.legend(loc='lower right')
plt.title('Spectral Centroid Mean vs Spectral Bandwidth Mean - Classical vs Pop')
plt.show()
```

```python
plt.figure(figsize=(12, 8), dpi=80)
sns.boxplot(data=data,x='label',y='chroma_stft_mean',
           order=['metal','hiphop','disco','pop','reggae','rock','country','blues','jazz','classical'])
plt.title('Chroma Frequency Mean by Genre')
plt.show()
```

```python
plt.figure(figsize=(12, 8), dpi=80)
mask = np.triu(np.ones_like(data.corr(), dtype=np.bool))
sns.heatmap(data.corr(), mask=mask, cmap='mako')
plt.show()
```

```python
plt.figure(figsize=(12, 8), dpi=80)
sns.boxplot(data=data,x='label',y='mfcc2_mean',
           order=['classical','country','jazz','blues','rock','reggae','disco','metal','hiphop','pop'])
plt.title('Mel-Frequency Cepstral Coefficient 2 Mean by Genre')
plt.show()
```
